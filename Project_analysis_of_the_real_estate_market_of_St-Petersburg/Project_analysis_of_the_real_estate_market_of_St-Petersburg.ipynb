```python
import pandas as pd
pd.set_option('display.float_format', '{:,.2f}'.format)

import numpy as np

from io import BytesIO

import requests

from matplotlib import pyplot as plt

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

import plotly.express as px
import plotly.graph_objects as go
```

## Функции

```python
def general_info(df):
    display(df.head())
    print('Общая информация:')
    print(df.info())
    print('В таблице {} столбцов и {} строк'.format(df.shape[1], df.shape[0]))
    print('Kоличество пропусков:')
    omssns = df.isna().sum()
    omissions = pd.DataFrame(data={'column': omssns.index, 
                                   'omissions': [x for x in omssns],
                                   '%-omissions': ['{:.2%}'.format(x/df.shape[0]) for x in omssns]})
    display(omissions.sort_values(by='omissions', ascending=False))
    print('Kоличество явных дубликатов:', df.duplicated().sum())
```

```python
# функция возвращает столбец с названиями населенного пункта
def name_rename(string):
    string = string.split()
    rez = []
    for s in string:
        if not s.islower(): #все названия начинаются с заглавной буквы, 
                            # поэтому отбрасываю все что состоит из строчных букв
            rez.append(s)
    return ' '.join(rez)
```

```python
# функция заполняет пропуски в столбце медианой в зависимости от выбранного столбца
def fillna_median(col_fill, col):
    df[col_fill].fillna(df.groupby([col])[col_fill].transform('median'), inplace=True)
```

```python
# функция строит матрицу корреляции для выбранного столбца
def corr_plot(col):
    plt.figure(figsize=(9,9))
    sns.heatmap(df.corr()[[col]].sort_values(by=col, ascending=False), square=True, annot=True, fmt='.2f')
    plt.show()
```

```python
# функция строит зависимость от населенного пункта
def loc_plot(col):
    plt.figure(figsize=(9,9))
    sns.lineplot(data=df, y=col, x='district')
    plt.xticks([])
    plt.title('Заивисмость '+col+' от населенного пункта')
    plt.show()
```

```python
def hist_plot(col, range=(20, 120), xl=['общая площадь', 'кв.м']):
    sns.displot(data=df, x=col, kde=True)
    plt.xlim(range)
    plt.xlabel(xl[0]+' '+xl[1])
    plt.title('Гистограмма по столбцу '+xl[0])
    plt.axvline(x=np.quantile(df[col], 0.25), color='g', linestyle='dashed', 
                label='0.25-квантиль: '+str(np.quantile(df[col], 0.25)))
    plt.axvline(x=np.quantile(df[col], 0.5), color='r', linestyle='dashed', label='медиана: '+str(np.quantile(df[col], 0.5)))
    plt.axvline(x=df[col].mean(), color='m', linestyle='dashed', label='среднее: '+str(round(df[col].mean())))
    plt.axvline(x=np.quantile(df[col], 0.75), color='y', linestyle='dashed', 
                label='0.75-квантиль: '+str(np.quantile(df[col], 0.75)))
    plt.legend()
    plt.show()
```

```python
def price_scatter(df_good, col, yl=['общая площадь', 'кв.м']):
    plt.figure(figsize=(8,8))
    #sns.scatterplot(data=df_good, x='last_price', y=col)
    sns.lmplot(data=df_good, x='last_price', y=col, line_kws={'color': 'r'}, aspect=1.5)
    plt.xlabel('цена, млн')
    plt.ylabel(yl[0]+', '+yl[1])
    plt.title('влияние на цену переменной '+yl[0])
    plt.text(20, df_good[col].mean()*0.4, 'Коэффициент корреляции: '+str(round(df_good['last_price'].corr(df_good[col]), 2)), 
             color='r')
    plt.show();
```

### Переименование столбцов

```python
#заменяю название столбца cityCenters_nearest на city_centers_nearest 
df = df.rename(columns={'cityCenters_nearest': 'city_centers_nearest',  
                        'parks_around3000': 'parks_around_3000', 
                        'ponds_around3000': 'ponds_around_3000'})
```

### Обработка пропусков


#### locality_name и floors_total

В этих столбцах менее 1% данных, удалю их.

```python
df.dropna(subset=['locality_name', 'floors_total'], inplace=True)
```

#### Добавление столбца с районами

Я заметила, что в названии населенного пункта встречаются неявные дубликаты.
<br>Все названия записаны с большой буквы, а типы с маленбкой.
<br>Уберу тип населенного пункта и запишу названия в новый столбец.

```python
df['new_locality_name'] = df['locality_name'].apply(name_rename)
```

Теперь добавлю столбец с районом из стороннего файла.

```python
df = df.merge(district, how='left', on='new_locality_name')
```

#### balcony, parks_around_3000, ponds_around_3000

Предполагаю, что наличие пропусков в этих столбцах, обьясняется тем, что таких особенностей у квартиры нет, значит пропуски можно заменить нулями.

```python
for col in ['balcony', 'parks_around_3000', 'ponds_around_3000']:
    df[col].fillna(0, inplace=True)
```

#### parks_nearest, ponds_nearest

Проверю, связанны ли эти пропуски с отсутствием парков и водоемов рядом:

```python
print('Количество пропусков, не связанное с отсутствием парков: {}'.format(
    len(df.query('parks_around_3000!=0 & parks_nearest.isna()'))))
```

```python
print('Количество пропусков, не связанное с отсутствием водоемов: {}'.format(
    len(df.query('ponds_around_3000!=0 & ponds_nearest.isna()'))))
```

Так как парков и водоемов нет в радиусе 3000м, заполню пропуски значением 3001м

```python
for col in ['parks_nearest', 'ponds_nearest']:
    df[col].fillna(3001, inplace=True)
```

Пропуски остались, дозаполню  по району

```python
fillna_median('days_exposition', 'district')
```

### Добавление столбцов с днем недели, месяцем и годом публикации объявлений

Добавлю день недели методом dt.weekday, он возвращает номер дня недели: понедельнику соответствует 0, а воскресенью 6, поэтому потом заменю цифры на названия дня недели.

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
```

```python
def for_name_weekday(weekday):
    dict_name_weekday = {0: 'понедельник', 1: 'вторник', 2: 'среда', 
                         3: 'четверг', 4: 'пятница', 5: 'суббота', 6: 'воскресенье'}
    return dict_name_weekday[weekday]
```

```python
df['weekday'] = df['weekday'].apply(for_name_weekday)
```

Чтобы получить год и месяц, я помещу время в класс DatetimeIndex и применю к нему атрибуты year и month

```python
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year
```

```python
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month
```

### Добавление столбца с категоризацией этажа квартиры

Разделю этажи квартиры по вариантам:
- первый
- последний
- другой

### Добавление столбцов с соотношением жилой и общей площади, а также с отношением площади кухни к общей

```python
#столбец с отношение жилой площади к общей
df['ratio_living_to_total'] = df['living_area'] / df['total_area']
```

```python
#столбец с оттношением площади кухни к общей
df['ratio_kitchen_to_total'] = df['kitchen_area'] / df['total_area']
```

### Исследование квартир в центральной зоне СПб

Создам df квартир в центре

```python
df_centre = df_spb.query('city_centers_nearest <= 10')
```

Выведу гистограммы следующих параметров:
- площадь
- цена
- число комнат
- высота потолков

```python
dict_col = {'total_area': (10, 200), 'last_price': (0, 50), 'rooms': (0, 8),
            'ceiling_height': (2.25, 4)}
dict_col_x = {'total_area': 'общая площадь, кв.м', 'last_price': 'цена, млн', 
              'rooms': 'количество комнат, шт', 'ceiling_height': 'высота потолков, м'}
for col in dict_col:
    ax = df_centre[col].plot(kind='hist', label=col+'_centre', bins=50, range = dict_col[col], histtype='step',
                            legend=True, figsize=(10,10))
    ax.axvline(x=np.quantile(df_centre[col], 0.25), color='lime', linestyle='dashed', 
                label='0.25-квантиль для центра: '+str(np.quantile(df_centre[col], 0.25)))
    ax.axvline(x=np.quantile(df_centre[col], 0.5), color='lightcoral', linestyle='dashed', label='медиана для центра: '+str(np.quantile(df_centre[col], 0.5)))
    ax.axvline(x=df_centre[col].mean(), color='plum', linestyle='dashed', label='среднее для центра: '+str(round(df_centre[col].mean())))
    ax.axvline(x=np.quantile(df_centre[col], 0.75), color='darkkhaki', linestyle='dashed', 
                label='0.75-квантиль для центра: '+str(np.quantile(df_centre[col], 0.75)))
    df[col].plot(kind='hist', label=col+'_df', bins=50, range = dict_col[col], ax=ax,
                histtype='step', legend=True)
    ax.axvline(x=np.quantile(df[col], 0.25), color='darkgreen', linestyle='dashed', 
                label='0.25-квантиль: '+str(np.quantile(df[col], 0.25)))
    ax.axvline(x=np.quantile(df[col], 0.5), color='darkred', linestyle='dashed', label='медиана: '+str(np.quantile(df[col], 0.5)))
    ax.axvline(x=df[col].mean(), color='purple', linestyle='dashed', label='среднее: '+str(round(df[col].mean())))
    ax.axvline(x=np.quantile(df[col], 0.75), color='gold', linestyle='dashed', 
                label='0.75-квантиль: '+str(np.quantile(df[col], 0.75)))
    ax.legend()
    ax.set_xlabel(dict_col_x[col])
    ax.set_ylabel('частота')
    plt.show()
```
